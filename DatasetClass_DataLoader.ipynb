## Implementing Breast Cancer problem with Dataset class and DataLoader 

```python
import torch
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train).float()
X_test_tensor = torch.from_numpy(X_test).float()
y_train_tensor = torch.from_numpy(y_train).float()
y_test_tensor = torch.from_numpy(y_test).float()
```

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class CustomDataset(Dataset):
    def __init__(self,features,labels):
        self.features = features
        self.labels = labels
    def __len__(self):
        return self.features.shape[0]
    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
train_dataset = CustomDataset(X_train_tensor,y_train_tensor)
test_dataset = CustomDataset(X_test_tensor,y_test_tensor)
```

```python
train_loader = DataLoader(train_dataset,batch_size=32,shuffle=True)
test_loader = DataLoader(test_dataset,batch_size=32,shuffle=True)
```

```python
import torch.nn as nn 

class SimpleNN(nn.Module):
    def __init__(self,input_features):
        super().__init__()
        self.linear1 = nn.Linear(input_features,3)
        self.relu1 = nn.ReLU()
        self.linear2 = nn.Linear(3,3)
        self.relu2 = nn.ReLU()
        self.output = nn.Linear(3,1)
        self.sigmoid = nn.Sigmoid()
    def forward(self,features):
        x = self.linear1(features)
        x = self.relu1(x)
        x = self.linear2(x)
        x = self.relu2(x)
        x = self.output(x)
        x = self.sigmoid(x)
        return x
```

```python
learning_rate = 0.001
num_epochs = 100
```

```python
model = SimpleNN(input_features=X_train_tensor.shape[1])
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)
loss_function = nn.BCELoss()
```

```python
for epoch in range(num_epochs):
    for batch_features,batch_labels in train_loader:
        y_pred = model(batch_features)
        loss = loss_function(y_pred.squeeze(),batch_labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Epoch : {epoch+1} , Loss : {loss.item()}')
```

```python
model.eval()
accuracy_list = []
with torch.no_grad():
    for batch_features,batch_labels in test_loader:
        y_pred = model(batch_features)
        y_pred = (y_pred.squeeze() >= 0.5).float()
        accuracy = (y_pred == batch_labels).float().mean()
        accuracy_list.append(accuracy.item())
    print(f'Accuracy : {sum(accuracy_list)/len(accuracy_list) * 100:.2f}%')
overall_accuracy = sum(accuracy_list)/len(accuracy_list)
print(f'Overall Test Accuracy: {overall_accuracy * 100:.2f}%')
```
